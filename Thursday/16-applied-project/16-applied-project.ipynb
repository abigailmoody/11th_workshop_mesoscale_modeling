## Effect of solvent quality on polymer structure

Solvent quality influences the structure of a polymer chain and, in turn, its
dynamics. Poor solvents cause the chain to collapse into a dense globule, while
good solvents cause the chain to swell. The scaling exponent $\nu$ for the
radius of gyration $R_{\rm g}$ with chain length $N$ is defined by
$R_{\rm g} \sim N^\nu$. For a poor solvent, $\nu \approx 0.33$; for an ideal
solvent $\nu = 0.5$; and for a good solvent, $\nu \approx 0.6$.

A top-down coarse-grained model for polymers in varying quality solvents can be
constructed as a variation on the Kremer-Grest model. We will represent the
polymer as a linear bead-spring chain with the same Kremer-Grest FENE bonds
between beads, but now, the beads will interact through a
[perturbed Lennard-Jones potential](https://azplugins.readthedocs.io/en/stable/module-azplugins-pair.html#hoomd.azplugins.pair.PerturbedLennardJones)
that has an attractive tail whose strength can be tuned through the parameter
$\lambda$ instead of the purely repulsive WCA potential. When $\lambda = 0$,
this potential is the same as WCA; when $\lambda = 1$, it is the Lennard-Jones
potential; and when $0 < \lambda < 1$, it has an attraction strength less than
the Lennard-Jones potential.

We would like estimate the value of $\lambda$ for which this model behaves like
an *ideal* chain. Conduct simulations of a few different polymer chain lengths
and a few different values of $\lambda$ to create scaling plots of $R_{\rm g}$
and extract the exponent $\nu$. Use the values of $\nu$ you calculated as a
function of $\lambda$ to determine the value for which $\nu = 0.5$.

## Block copolymers

Block copolymer micelles are generally formed by the self-assembly of either
amphiphilic or oppositely charged copolymers in aqueous medium. These micelles
can be spherical, cylindrical, or adopt more complex self-assembled
morphologies. Micelles in particular have application in drug delivery or as
building block for hierarchical self-assembled structures. Modify the
Kremer-Grest model to make a simple AB block copolymer with one hydrophobic
block (FENE+LJ) of length $N_A$ and one hydrophilic block (FENE+WCA) of length
$N_B$. Simulate dilute solutions of polymers with varying degree of
$f_A=N_A/(N_A+N_B). Find the range of $f_A$ and concentration that forms
micelles. Use freud to calculate the size of each micelle by using the "Cluster"
module. Compute the average radius of gyration of the micelles. Compute the
diffusion coefficient of micelles as function of their size. 

## Structure of a polymer in shear flow

In previous exercises, you analyzed the structure (radius of gyration) of a
Gaussian chain at equilibrium and performed MPCD simulations to simulate shear
flow. The objective of this project is to examine how the structure of the
Gaussian chain is perturbed in shear flow, which is a scenario that is
commonly encountered in the rheological characterization of complex fluids.
Perform simulations of a 50-mer Gaussian polymer chain under different flow
conditions to vary the shear rate. Characterize the relative deformation along
the direction of flow by computing the gyration tensor (see Eqs. 17 and 18 in
[10.1021/ma101836x](https://doi.org/10.1021/ma101836x)). Determine the value of
the Weissenberg number ${\rm Wi} = \dot \gamma \tau_{\rm R}$, where $\dot\gamma$
is the shear rate and $\tau_{\rm R}$ is the relaxation time of the chain,
at which the chains being to elongate. If time permits, re-perform the analysis
for different chain lengths and also characterize the alignment of the polymers
with the flow via Eq. 19 in the same reference.

## Drag coefficient of a cube

Nanoparticles with a variety of shapes can be synthesized, but the transport
properties of these anisotropic particles can be less well-understood than
those of spherical particles. Evaluate the drag coefficient for a cube from an
MPCD sedimentation simulation. Model one cube having edge length 6 using a
discretized representation for each face as described in
[Wani et al.](https://doi.org/10.1039/D4SM00271G). The surface particles should
be bonded to their nearest neighbors and to a central particle. The same
finite-size correction used for the sedimentation velocity of a sphere will
also work for the cube. Compare your simulation results to the empirical
correlation given by Eq. 9 of the same paper referenced above.
